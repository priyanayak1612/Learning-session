```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import random
```

```python
#Drawing a graph using the functions above./using matplotlib.pyplot library to draw a graph/first distrubution
incomes = np.random.normal(50000,20000,10000)
plt.hist(incomes,50)
plt.show()
```

```python
#Line graph
a1 = np.random.normal(12,25,10)
a2 = np.random.normal(12,25,10)
plt.plot(a1,a2)
plt.show()
```

```python
#Multiple graphs in one graph
a1 = np.array([1,2,3])
a2 = np.array([[7,8,9],[10,11,12]])
a3 = np.random.normal(12,13,2)
plt.plot(a2,a3)
plt.plot(a1,[4,5,3])
plt.show()
```

```python
gr2 = np.array([1,2,3])
gr3 = np.array([[7,8,9],[10,11,12]])
gr3y = np.random.normal(12,13,2)
plt.plot(gr3,gr3y)
plt.plot(gr2,[4,5,6])
plt.show()
```

```python
#Pie chart
pie1 = np.array([10,20,30])
pie2 = [10,20,30,40]
plt.pie(pie1)
```

```python
pie1 = np.array([10,20,30])
pie2 = [10,20,30,40]
plt.pie(pie2)
```

```python
#Bar chart
bar = [10,22]
bar_w = np.array([2,1])
plt.bar(bar,20,bar_w,1,align = 'center')
```

```python
#Scatter plot
x = np.random.normal(100,200,500)
y = np.random.normal(300,400,500)
plt.scatter(x,y)
```

```python
#Histogram
hist = np.random.normal(22,55)
plt.hist(hist)
```

```python
#Boxplot
p = np.random.normal(45,68,90)
plt.boxplot(p,60)
```

```python
#Random distributions
a = np.random.normal(2,5,7)
b = np.random.noncentral_f(50,44,3)
c = np.random.pareto(44,5)
d = np.random.permutation(33)
print(a,end ='*')
print(b,end ='**')
print(c,end ='***')
print(d,end ='****')
```
